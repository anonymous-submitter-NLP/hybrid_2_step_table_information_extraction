Get the Path to the results excel and the folder of all tables and rename to FALSE_RIGHT depending on whether the recall is 0 or 1 

```python
import pandas as pd
import glob
import os
```

```python
path_to_results = "/Users/hendrikweichel/projects/ReMeDi/remedi/Tests/test_results/test_rule_based_extraction/tables_prior_table_row_name_regular_expression/0_resultsprior_table_row_name_regular_expression.xlsx"
path_to_tables = "/Users/hendrikweichel/projects/ReMeDi/remedi/Tests/test_results/test_rule_based_extraction/tables_prior_table_row_name_regular_expression/"
```

```python
df = pd.read_excel(path_to_results)
df.head()
```

```python
def rename_file(file_path, new_name):
    directory = os.path.dirname(file_path)
    new_path = os.path.join(directory, new_name)
    os.rename(file_path, new_path)
    return new_path

# # Example usage
# new_file_path = rename_file(path_to_results, "0_resultsprior_table_row_name_regular_expression.xlsx")
# print(f"File renamed to: {new_file_path}")
```

```python
df[df["recall"] == 0].apply(lambda row: rename_file(os.path.join(path_to_tables, f'{row["report"]}_{row["emission_type"]}.xlsx'), f'false_{row["report"]}_{row["emission_type"]}.xlsx'), axis=1)
```

```python
df[df["recall"] == 1].apply(lambda row: rename_file(os.path.join(path_to_tables, f'{row["report"]}_{row["emission_type"]}.xlsx'), f'true_{row["report"]}_{row["emission_type"]}.xlsx'), axis=1)
```
